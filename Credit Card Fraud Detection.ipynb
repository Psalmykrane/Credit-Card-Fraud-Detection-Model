## Credit Card Fraud Detection Model

#### Import Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Load dataset to a pandas dataframe
card_data = pd.read_csv('creditcard.csv')
```

#### Data Exploration and Exploratory Analysis

```python
# first five rows of the dataset
card_data.head()
```

```python
# last five rows of the data frame
card_data.tail()
```

```python
# dataset information
card_data.info()
```

```python
# Checking number of missing values in each column
card_data.isnull().sum()
```

```python
# checking distribution of legit transactions and fraudulent transactions under the class column
card_data['Class'].value_counts()
```

#### 0 --> Legit transactions
#### 1 --> Fraudulent transactions

##### From the above, this dataset is highly unbalanced

```python
# Separating the data for analysis
legit = card_data[card_data.Class == 0]
fraud = card_data[card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
card_data.groupby('Class').mean()
```

#### Under-sampling

#### Build a sample dataset containing a similar distribution of normal transaction and fraudulent transaction
#### Number of Fraudulent transactions --> 492

```python
legit_sample = legit.sample(n = 492)
```

#### concatenating two dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
# first five rows of new dataframe
new_dataset.head()
```

```python
# last five rows of dataframe
new_dataset.tail()
```

```python
# checking the number of values
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

#### Splitting the data into features and target data

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

#### Splitting into Training and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#### Model Training

```python
# Building Logistic Regression Model
model = LogisticRegression()
```

```python
# Training the logistic regression model with training data
model.fit(X_train, Y_train)
```

#### Model Evaluation

```python
# Accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of model on Training data:', training_data_accuracy)
```

###### Model is 94.8% accurate with Training data

```python
# Accuracy score on training data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of model on Test data:', test_data_accuracy)
```

###### Model is 90.9% accurate with Test data
